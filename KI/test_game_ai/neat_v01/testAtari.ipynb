```python
import gym

```

```python

env = gym.make("MountainCar-v0")
# env = gym.make("Breakout-v0")
```

```python

observation = env.reset()

print(env.observation_space)
print(env.action_space)

```

```python
done = False
while not done:
    observation, reward, done, info = env.step(env.action_space.sample())
    print(reward)

    env.render()

```

```python
env.close()
```
